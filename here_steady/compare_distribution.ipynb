```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
```

```python
df1 = pd.read_pickle('dis_seasonal.pkl')
df1.T[1:]
```

```python
df1.T[1:].mean().mean()
```

```python
df1.T[1:].std().mean()
```

```python
nrows = 20
ncols = 24

fig, ax = plt.subplots()
cp = ax.imshow(df1.T[1:], cmap="cool", origin="lower", vmin=0)

# optionally add grid
ax.set_xticks(np.arange(ncols+1)-0.5, minor=True)
ax.set_yticks(np.arange(nrows+1)-0.5, minor=True)
ax.set_xlabel('Time of day (hour)')
ax.set_ylabel('Region ID')
# ax.grid(which="minor")
ax.tick_params(which="minor", size=0)
fig.colorbar(cp)
plt.show()
```
