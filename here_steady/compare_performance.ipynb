```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams.update({'font.size': 18})
```

```python
flow_20_true = pd.read_pickle('flow_20_true.pkl')
flow_20_pre = pd.read_pickle('flow_20.pkl')
flow_20_old = pd.read_pickle('flow_20_old.pkl')
plt.plot(flow_20_true.T.sum(), label = 'AT',color='c')
plt.plot(flow_20_pre.T.sum(), label = 'TP-WF',color='b')
plt.plot(flow_20_old.T.sum(), label = 'TR-NF',color='m')
plt.ylim(0,400)
plt.xlabel('Time of day')
plt.ylabel('Traffic flow')
plt.legend()
plt.grid()
```

```python
def smape(a, f):
    return 1/len(a) * np.sum(2 * np.abs(f-a) / (np.abs(a) + np.abs(f))*100)
```

```python
smape(flow_20_true, flow_20_pre).mean()
```

```python
smape(flow_20_true, flow_20_old).mean()
```

```python
import numpy as np
import scipy.stats as st

gfg_data =  smape(flow_20_true, flow_20_pre)

# create 95% confidence interval
st.t.interval(alpha=0.95, df=len(gfg_data)-1,
              loc=np.mean(gfg_data),
              scale=st.sem(gfg_data))
```

```python
gfg_data =  smape(flow_20_true, flow_20_old)

# create 95% confidence interval
st.t.interval(alpha=0.95, df=len(gfg_data)-1,
              loc=np.mean(gfg_data),
              scale=st.sem(gfg_data))
```
